```python
Corr_dir = '/global/cscratch1/sd/huikong/obiwan_Aug/repos_for_docker/obiwan_out/obiwan_corr/corr_output/'
fn1 = 'WEIGHT_SYSTOT_chunk22.out'#chunk22 using WEIGHT_SYSTOT
```

```python
%pylab inline
import matplotlib.pyplot as plt
def corr_plot(Corr_dir,fn):
    plot_title = fn[:-4]
    dat = np.loadtxt(Corr_dir+fn).transpose()
    plt.gca().set_xscale("log", nonposx='clip')
    plt.errorbar(dat[0],dat[1]*dat[0],dat[0]*dat[2])
    plt.xlabel(r'$\theta$'+'(Degree)')
    plt.ylabel(r'$\theta$'+'*w('+r'$\theta$'+')')
    plt.title(plot_title)
```

```python
#chunk22 using WEIGHT_SYSTOT
corr_plot(Corr_dir,fn1)
```

```python
fn2 = 'WEIGHT_SYSTOT_chunk21.out'
#chunk21 using WEIGHT_SYSTOT
corr_plot(Corr_dir,fn2)
```

```python
#the uniform randoms that is used to make chunk21 WEIGHT_SYSTOT
fn3 = 'uniform_chunk21.out'
corr_plot(Corr_dir,fn3)
```

```python
dat2 = np.loadtxt(Corr_dir+fn2).transpose()
dat3 = np.loadtxt(Corr_dir+fn3).transpose()
plt.errorbar(dat2[0],dat2[1]*dat2[0],dat2[0]*dat2[2],color = 'red',label = 'WEIGHT_SYSTOT')
plt.errorbar(dat3[0],dat3[1]*dat3[0],dat3[0]*dat3[2],color = 'blue', label = 'uniform')
plt.legend()
plt.title('comparison between chunk21 randoms and chunk21 weight')
```
